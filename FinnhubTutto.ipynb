```python
import os
import shutil

# Source directory containing duplicate folders
source_dir = '/content/FinnhubData'

# Destination directory for merged folders
destination_dir = '/content/FinnhubData/destination'

# Iterate over all folders in the source directory
for folder_name in os.listdir(source_dir):
    folder_path = os.path.join(source_dir, folder_name)

    # Check if the item in the source directory is a folder
    if os.path.isdir(folder_path):
        # Check if the folder already exists in the destination directory
        if os.path.exists(os.path.join(destination_dir, folder_name)):
            # Move all files from the duplicate folder to the destination folder
            for file_name in os.listdir(folder_path):
                file_path = os.path.join(folder_path, file_name)
                destination_path = os.path.join(destination_dir, folder_name, file_name)
                shutil.move(file_path, destination_path)
            # Remove the empty duplicate folder
            os.rmdir(folder_path)
        else:
            # If the folder doesn't exist in the destination directory, move it there
            shutil.move(folder_path, destination_dir)
```

```python
import os
import shutil

def merge_quarterly_data(source_dir, destination_dir):
    for item in os.listdir(source_dir):
        if os.path.isdir(os.path.join(source_dir, item)):
            year = item.split('.')[0]
            year_destination_dir = os.path.join(destination_dir, year)
            if not os.path.exists(year_destination_dir):
                os.makedirs(year_destination_dir)
            quarterly_source_dir = os.path.join(source_dir, item)
            files_to_move = os.listdir(quarterly_source_dir)
            for file_name in files_to_move:
                source_file_path = os.path.join(quarterly_source_dir, file_name)
                destination_file_path = os.path.join(year_destination_dir, file_name)
                if os.path.exists(destination_file_path):
                    # If file with same name exists in destination, rename the file
                    destination_file_path = os.path.join(year_destination_dir, f"{item}_{file_name}")
                shutil.move(source_file_path, destination_file_path)
            # Remove the empty quarterly folder
            os.rmdir(quarterly_source_dir)

if __name__ == "__main__":
    source_directory = "/content/FinnhubData"
    destination_directory = "/content/ConsolidatedData"
    merge_quarterly_data(source_directory, destination_directory)
```
